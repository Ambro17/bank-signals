```
#@title Carga del dataset. Entrada/Salida
dataset = pd.read_csv('/content/RNA-IA-2020/Dataset-Banco.csv')
X = dataset.iloc[:, [6,5,8,12,10,11]].values  # Datos entrada (Edad,Sexo,Balance,SalarioEstimado,PoseeTarjetaCredito,EsMiembroActivo)
Y = dataset.iloc[:, [13]].values              # Dato salida (Abandono)
```

```
#@title Codificar datos categóricos - Sexo

from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
```

```
#@title Split Training and Testing Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```
#@title Normalizacion de variables
from sklearn.preprocessing import StandardScaler
standarScaler = StandardScaler()
X_train = standarScaler.fit_transform(X_train)
X_test = standarScaler.transform(X_test) 
```

```
#@title Importar Keras y sus librerias
import keras
from keras.models import Sequential 
from keras.layers import Dense 
from keras.layers import Dropout 
```

```
#@title Init
# Inicializar la RNA y confeccion de capas - (6 neuronas de entrada, 1 capa oculta de 8 neuronas , 1 capa de salida)

RNA = Sequential() 

# Añadir las capas de entrada y primera capa oculta

RNA.add(Dense(units = 8, kernel_initializer = "uniform",  activation = "relu", input_dim = 6))
RNA.add(Dropout(0.1))

# Añadir la capa de salida
RNA.add(Dense(units = 1, kernel_initializer = "uniform",  activation = "sigmoid"))

# Compilar la RNA 
RNA.compile(optimizer = "sgd", loss = "mean_squared_error", metrics = ["accuracy"])


```

```
#@title Train
# Entrenamos la RNA con el Conjunto de Entrenamiento
RNA.fit(X_train, Y_train,  batch_size = 15, epochs = 300)
```

```
#@title Predict
#Predicción de los resultados con el Conjunto de Testing
#Tomamos que aquellas probabilidades mayores a 0.5 es efectivamente un abandono
Y_pred  = RNA.predict(X_test)
Y_pred = (Y_pred>0.5)
```

```
#@title Result metrics
# Crear matriz de confusión
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print("Exactitud = " + str((cm[0][0]+cm[1][1])/cm.sum()))
print("Precision = " + str(cm[0][0]/(cm[0][0]+cm[1][0])))
print("Recuperacion = " + str(cm[0][0]/(cm[0][0]+cm[0][1])))
```
